```python
# How many days on average are customers reallocated to a different node?

avg_reallocation_days = (cn_df1['end_date'] - cn_df1['start_date']).mean().days

print(f"\nAverage Reallocation Days: {avg_reallocation_days} days")
```

cn_df1['end_date'] - cn_df1['start_date']: Calculates the time difference between the 'end_date' and 'start_date' columns for each row in the DataFrame cn_df1. This results in a Series of timedelta values.

.mean(): Calculates the mean (average) of the timedelta values, representing the average time difference between 'end_date' and 'start_date'.

.days: Extracts the number of days from the resulting average timedelta value.

The result, stored in avg_reallocation_days, is the average number of days customers are reallocated to a different node.

print(f"\nAverage Reallocation Days: {avg_reallocation_days} days"): Prints a formatted string that includes the average reallocation days.
So, when you run this code, it will output the average number of days customers spend in a reallocation to a different node.

```python
# What is the median, 80th and 95th percentile for this same reallocation days metric for each region?

cn_df1['reallocation_days'] = (cn_df1['end_date'] - cn_df1['start_date']).dt.days

reallocation_days_percentiles = cn_df1.groupby('region_id')['reallocation_days'].quantile([0.5, 0.8, 0.95]).reset_index()

reallocation_days_percentiles
```

cn_df1['reallocation_days']: Creates a new column 'reallocation_days' in the DataFrame cn_df1.

(cn_df1['end_date'] - cn_df1['start_date']).dt.days: Calculates the time difference between 'end_date' and 'start_date' for each row in cn_df1, then extracts the number of days and assigns it to the new 'reallocation_days' column.

cn_df1.groupby('region_id'): Groups the DataFrame cn_df1 by the 'region_id' column.

['reallocation_days'].quantile([0.5, 0.8, 0.95]): Calculates the specified percentiles (median, 80th percentile, and 95th percentile) for the 'reallocation_days' column within each group.

.reset_index(): Resets the index of the resulting Series and converts it back into a DataFrame. This is done to have a clearer structure with 'region_id' as a separate column.

The result, stored in reallocation_days_percentiles, is a DataFrame with three columns: 'region_id', 'reallocation_days', and the specified percentiles.

```python
# What is the unique count and total amount for each transaction type?
transaction_stats = ct_df1.groupby('txn_type').agg({'customer_id': 'nunique', 'txn_amount': 'sum'}).reset_index()

transaction_stats.columns = ['txn_type', 'unique_count', 'total_amount']

print("\nTransaction Statistics:")

transaction_stats
```

ct_df1.groupby('txn_type'): Groups the DataFrame ct_df1 by the 'txn_type' column.

.agg({'customer_id': 'nunique', 'txn_amount': 'sum'}): Applies aggregation functions to the grouped data. It calculates the unique count of 'customer_id' and the sum of 'txn_amount' for each group.

.reset_index(): Resets the index of the resulting DataFrame.

Assigns new column names to the DataFrame. The original column names, 'customer_id' and 'txn_amount', are replaced with 'unique_count' and 'total_amount', respectively.

```python
# What is the average total historical deposit counts and amounts for all customers?

avg_deposit_counts = ct_df1[ct_df1['txn_type'] == 'deposit'].groupby('customer_id')['txn_date'].count().mean()
avg_deposit_amount = ct_df1[ct_df1['txn_type'] == 'deposit'].groupby('customer_id')['txn_amount'].sum().mean()

print(f"\nAverage Deposit Counts: {avg_deposit_counts}")
print(f"Average Deposit Amount: {avg_deposit_amount}")
```

ct_df1[ct_df1['txn_type'] == 'deposit']: Filters the DataFrame ct_df1 to include only rows where the 'txn_type' is 'deposit'.

.groupby('customer_id')['txn_date'].count(): Groups the filtered DataFrame by 'customer_id' and calculates the count of 'txn_date' (number of historical deposit counts) for each customer.

.mean(): Calculates the mean (average) of the deposit counts across all customers.

ct_df1[ct_df1['txn_type'] == 'deposit']: Similar to the previous line, filters the DataFrame to include only rows where the 'txn_type' is 'deposit'.

.groupby('customer_id')['txn_amount'].sum(): Groups the filtered DataFrame by 'customer_id' and calculates the sum of 'txn_amount' (total historical deposit amounts) for each customer.

.mean(): Calculates the mean (average) of the deposit amounts across all customers.

```python
# For each month - how many Data Bank customers make more than 1 deposit and either 1 purchase or 1 withdrawal in a single month?

monthly_activity = (ct_df1[(ct_df1['txn_type'] == 'deposit') &
                           (ct_df1['customer_id'].isin(ct_df1[ct_df1['txn_type']
                                                              .isin(['purchase', 'withdrawal'])].customer_id.unique()))]
                    .groupby(ct_df1['txn_date'].dt.to_period("M"))
                    .agg({'customer_id': 'nunique'}).reset_index())
monthly_activity.columns = ['month', 'customers_count']
print("\nMonthly Activity:")
monthly_activity
```

ct_df1['txn_type'] == 'deposit': Filters the DataFrame to include only rows where the 'txn_type' is 'deposit'.

(ct_df1['customer_id'].isin(ct_df1[ct_df1['txn_type'].isin(['purchase', 'withdrawal'])].customer_id.unique())): Filters further to include only rows where the 'customer_id' is present in the list of customers who made at least one purchase or one withdrawal.

.groupby(ct_df1['txn_date'].dt.to_period("M")): Groups the filtered DataFrame by the month, using the 'txn_date' column and converting it to periods with a frequency of "M" (month).

.agg({'customer_id': 'nunique'}): Aggregates the number of unique customer IDs for each month.

.reset_index(): Resets the index of the resulting DataFrame.

Assigns new column names to the DataFrame. The original column names, 'txn_date' and 'customer_id', are replaced with 'month' and 'customers_count', respectively.

```python
# What is the closing balance for each customer at the end of the month?

ct_df1['month'] = ct_df1['txn_date'].dt.to_period("M")

closing_balance = (ct_df1.groupby(['customer_id', 'month'])
                   .apply(lambda x: x['txn_amount'] *
                          (-1 if x['txn_type'].isin(['purchase', 'withdrawal']).any() else 1))
                   .groupby(['customer_id', 'month'])
                   .sum().reset_index())

closing_balance.columns = ['customer_id', 'month', 'closing_balance']

print("\nClosing Balance:")

closing_balance
```

Creates a new column 'month' in the DataFrame ct_df1 containing the month (Period) corresponding to each transaction date.

ct_df1.groupby(['customer_id', 'month']): Groups the DataFrame by 'customer_id' and 'month'.

.apply(lambda x: x['txn_amount'] * (-1 if x['txn_type'].isin(['purchase', 'withdrawal']).any() else 1)): Applies a lambda function to calculate the adjusted transaction amounts. If any transaction in the group is a purchase or withdrawal, the transaction amount is negated; otherwise, it remains positive.

.groupby(['customer_id', 'month']): Groups the DataFrame again by 'customer_id' and 'month' after applying the lambda function.

.sum(): Calculates the sum of adjusted transaction amounts for each customer and month.

.reset_index(): Resets the index of the resulting DataFrame.

Assigns new column names to the DataFrame. The original column names are replaced with 'customer_id', 'month', and 'closing_balance', respectively.
