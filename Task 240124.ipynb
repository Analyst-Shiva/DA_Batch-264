```python
import matplotlib.pyplot as plt, seaborn as sns, numpy as np, pandas as pd
import matplotlib.gridspec as gridspec
import warnings
warnings.filterwarnings("ignore")

# Set a random seed for reproducibility
np.random.seed(42)

# Generate data
data_non_normal = np.random.exponential(size=(100,))

# Create a normal distribution for the second row
data_normal = np.random.normal(size=(100,))

# Create a DataFrame for the countplot
data_countplot = pd.DataFrame({'Category': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C']})

# Create a (3, 3) grid of axes
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

G = gridspec.GridSpec(3, 3)
axes_violin = plt.subplot(G[0, :])

# First row - Violin plot
sns.violinplot(x=data_non_normal, ax=axes_violin, orient='h')
axes_violin.set_title('Violin Plot (Non-normal Distribution)')

# Second row - Histogram, Density plot, Scatter plot
sns.histplot(data_normal, ax=axes[1, 0], kde=False, bins=20)
axes[1, 0].set_title('Histogram (Normal Distribution)')

sns.kdeplot(data_normal, ax=axes[1, 1])
axes[1, 1].axvline(np.mean(data_normal), color='r', linestyle='--', label='Mean')
axes[1, 1].axvline(np.median(data_normal), color='g', linestyle='--', label='Median')
axes[1, 1].axvline(np.mean(data_normal) + np.std(data_normal), color='b', linestyle='--', label='+1 Std Dev')
axes[1, 1].axvline(np.mean(data_normal) - np.std(data_normal), color='b', linestyle='--', label='-1 Std Dev')
axes[1, 1].axvline(np.mean(data_normal) + 2 * np.std(data_normal), color='purple', linestyle='--', label='+2 Std Dev')
axes[1, 1].axvline(np.mean(data_normal) - 2 * np.std(data_normal), color='purple', linestyle='--', label='-2 Std Dev')
axes[1, 1].axvline(np.mean(data_normal) + 3 * np.std(data_normal), color='orange', linestyle='--', label='+3 Std Dev')
axes[1, 1].axvline(np.mean(data_normal) - 3 * np.std(data_normal), color='orange', linestyle='--', label='-3 Std Dev')
axes[1, 1].set_title('Density Plot with Std Dev Lines')
axes[1, 1].legend()

sns.scatterplot(x=np.random.randn(100), y=np.random.randn(100), ax=axes[1, 2])
axes[1, 2].set_title('Scatter Plot')

# Third row - Exploded Pie Plot, Countplot, Butterfly Chart
explode = (0.1, 0, 0)
axes[2, 0].pie([30, 40, 30], explode=explode, labels=['A', 'B', 'C'], autopct='%1.1f%%', startangle=90)
axes[2, 0].set_title('Exploded Pie Plot')

sns.countplot(x='Category', data=data_countplot, ax=axes[2, 1])
axes[2, 1].set_title('Countplot')

butterfly_data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Value': [10, 20, 30]
})

sns.barplot(x='Value', y='Category', data=butterfly_data, ax=axes[2, 2], color='blue', label='Value')
sns.barplot(x=-butterfly_data['Value'], y='Category', data=butterfly_data, ax=axes[2, 2], color='orange', label='-Value')
axes[2, 2].set_title('Butterfly Chart')
axes[2, 2].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

Violin Plot (Non-normal Distribution):

A violin plot is created using Seaborn for a non-normal distribution generated with np.random.exponential.
Displayed in the first row, spanning all three columns.
Histogram (Normal Distribution):

A histogram is created using Seaborn for a normal distribution generated with np.random.normal.
Displayed in the second row, first column.
Density Plot with Standard Deviation Lines (Normal Distribution):

A density plot is created using Seaborn with vertical lines indicating the mean, median, and standard deviations.
Displayed in the second row, second column.
Scatter Plot:

A scatter plot is created using Seaborn with random data generated using np.random.randn.
Displayed in the second row, third column.
Exploded Pie Plot:

An exploded pie plot is created using Matplotlib for illustrating percentages with labels A, B, and C.
Displayed in the third row, first column.
Countplot:

A countplot is created using Seaborn for the 'Category' variable in the data_countplot DataFrame.
Displayed in the third row, second column.
Butterfly Chart:

A butterfly chart is created using Seaborn for positive and negative values in the butterfly_data DataFrame.
Displayed in the third row, third column.
The plt.tight_layout() function is used to improve the subplot layout, and plt.show() is used to display the entire figure.
